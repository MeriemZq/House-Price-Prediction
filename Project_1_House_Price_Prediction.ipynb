Slit the data set into train and test data

**train_test_split:** This function is part of the sklearn.model_selection module. It helps divide  dataset into two parts: one for training your machine learning model and the other for testing it.
**drop** will remove the specified feature which is the target that we want to predict ( axis = 1 => column)


```python
from sklearn.model_selection import train_test_split
x = data.drop(['median_house_value'],axis=1)
y = data['median_house_value']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

Joindre les x train data with the y train data ain a dataframe

```python
train_data= x_train.join(y_train)
```

```python
train_data.hist(figsize=(15,8))
```

We visualize the correlation between numeric features in our dataset using a heatmap:

Figure Size: The heatmap is sized at 15x8 inches for better visibility.

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include=[float, int]).corr(), annot=True, cmap='YlGnBu')
plt.show()
```

 will helps to reduce skewness and stabilize variance, making the data more normally distributed, which is often beneficial for modeling.

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
train_data.hist(figsize=(15,8))
```

Categories: Each unique category in the ocean_proximity column is listed, along with the count of how many entries belong to that category.
Distribution Insight: This output gives you insight into how many housing units are near the ocean, inland, or on an island. Understanding this distribution is crucial because proximity to the ocean can affect housing prices, desirability, and other features.

```python
train_data.ocean_proximity.value_counts()
```

With pd.get_dummies(train_data.ocean_proximity) it convert the values into 0 and 1 like if the house is near ocean its 1 and the others are 0 , then will join this new values into the train_data instead of ocean_proximity

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

feature Engineering : here we aim to create new features sounding intersting

```python
train_data['ratio_bedrooms'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include=[float, int]).corr(), annot=True, cmap='YlGnBu')
plt.show()
```

**Regression Linear model**


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = train_data.drop(['median_house_value'],axis=1)
y_train = train_data['median_house_value']
x_train_s = scaler.fit_transform(x_train)

reg = LinearRegression()
#Train the model with the training dataset
reg.fit(x_train_s,y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity).astype(int)).drop(['ocean_proximity'],axis=1)

test_data['ratio_bedrooms'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
x_test = test_data.drop(['median_house_value'],axis=1)
y_test = test_data['median_house_value']
```

```python
x_test_s = scaler.transform(x_test)
```

The model got 68,5% approximatly the y_test is not bad nore that good like if the actual y_test =50000$
he predict it like 52000

```python
reg.score(x_test_s,y_test)
```
